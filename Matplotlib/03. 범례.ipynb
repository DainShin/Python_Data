# 03. 범례

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
matplotlib.rcParams['font.size'] = 9 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스 글자가 깨지는 현상 해결 
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y,label='무슨 데이터')
plt.legend()  # 레전드를 통해 label을 출력
```

```python
plt.plot(x,y,label='무슨 데이터')
plt.legend(loc='upper right')
```

```python
plt.plot(x,y,label='무슨 데이터')
plt.legend(loc='lower left')  
```

```python
plt.plot(x,y,label='범례')
plt.legend(loc=(0.5, 0.5)) # x축, y축 (0~1사이)
```
