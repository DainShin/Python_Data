# 13. 원 그래프(기본)

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
matplotlib.rcParams['font.size'] = 9 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스 글자가 깨지는 현상 해결 
```

```python
values = [30, 25, 20 ,13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']

plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False)  # 소숫점 아래 첫번째 자리까지
plt.show()
```

```python
values = [30, 25, 20 ,13, 10, 2]
labels = ['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC']
# explode = [0.2, 0.1, 0, 0, 0,0]
explode = [0.05] *6   # 모든 데이터가 0.05만큼 틈이 생김

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels= labels, explode=explode)
plt.legend(loc=(1.2, 0.3))
plt.show()
```

```python
plt.pie(values, labels= labels, explode=explode)
plt.title('언어')  # 원그래프 제목
plt.legend(loc=(1.2, 0.3), title='언어별 선호도')
plt.show()
```
