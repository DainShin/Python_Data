# 2. 축

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic" # Windows
matplotlib.rcParams['font.size'] = 9 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글폰트 사용시, 마이너스 글자가 깨지는 현상 해결 
```

```python
x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family': 'HYGungSo-Bold', 'size':20}) # 개별 폰트 적요
```

```python
plt.plot(x,y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x,y)
plt.xlabel('X축', color='red') # x축은 빨간색
plt.ylabel('Y축', color='#00aa00')  # y축은 초록색
```

```python
plt.plot(x,y)
plt.xlabel('X축', color='red', loc='right') # left, center, right
plt.ylabel('Y축', color='#00aa00', loc='top')  #  top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```
